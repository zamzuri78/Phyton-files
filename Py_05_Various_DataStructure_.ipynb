```python
# LIST
# LIST IS A COLLECTION OF DATA VALUES / ELEMENTS
# LIST IS DENOTED BY USING [] / list()
# LIST IS MUTABLE [CHANGEABLE / MODIFIABLE / DELETION / UPDATION]
# LIST ALLOWS US TO HAVE DUPLICATE ELEMENTS

# CREATE A LIST

list1 = [111,22,33,54,51]
print(type(list1))

list2 = list(['a','b','c','d'])
print(type(list2))

list3 = []           # empty list
print(type(list3))

info = [101,"JOHM",78,21,True]
print(type(info))
```

```python
# indexing and slicing
list1 = [10,20,30,40,50,60,70,80]
print(list1)

print(list1[1])
print(list1[-1])
print(list1[1:6])
print(list1[1:6:2])
print(list1[::-1]) #mula dari belakang


# 10  20   30    40    50    60    70    80
# 0   1    2     3     4      5    6      7
# -8  -7   -6    -5    -4     -3   -2     -1
```

```python
#list is iterable
list1 = [10,20,30,40,50,60,70,80]
for i in list1:
    if i%3==0:
      print(i)
```

```python
#print the elements f list along with index numbers
list1 = [10,20,30,40,50,60,70,80]
for i in range(len(list1)): #len function tu nak kira berapa banyak data dalam list tu
    print(i,list1[i])


# Print the elements of list along with index numbers
list1 = [10,20,30,40,50,60,70,80]
# len() : used to count the number of elements in a List
print(f"jumlah data dalam list adalah =",len(list1))
```

```python
#enumerate()
for i in range(len(list1)):                   # range(8)  --> 0 1 2 3 4 5 6 7
  print(f'Index {i} : {list1[i]}')            #  i --> 0    list1[2]10
```

```python
# enumerate()  it will show elements along with index numbers
for x,e in enumerate(list1):
  print(f'Index {x} : {e}')
```

```python
# enumerate()  it will show elements along with index numbers
for i in enumerate(list1): #profesional programmer guna yang ini mereka tak assign variable pon trus ja pakai

  print(i)
```

```python
# c.) pop(indexNumber)
# used to delete an element from list
list1 = [10,20,30,40,50,60,70,80]
list1.pop(0)
print('After Pop :',list1)

# d.) pop()
# if no index number was given , it will delete the Last element
list1.pop()
print('After Pop :',list1)

# e.) remove(value)
# it will delete the specified value
list1.remove(20)
print('After Remove :',list1)

# list = [10,20,30,10,10,10,40,50]
# list.remove(10) ? which 10 it will delete
# it will delete the first occurence of 10
# list = [20,30,10,10,10,40,50]
```

```python
# f.) sort()
list1 = [67,32,19,80,12,17,45,26,46,78]
list1.sort()   # ascending order
print(list1)
```

```python
list1 = [67,32,19,80,12,17,45,26,46,78]
list1.sort(reverse=True)   # descending order
print(list1)
```

```python
# g.) reverse()
list1 = [67,32,19,80,12,17,45,26,46,78]
list1.reverse()  # [::-1]
print(list1)
```

```python
#tuple guna ()

# TUPLE
# TUPLE IS A COLLECTION OF DATA VALUES / ELEMENTS
# TUPLE IS DENOTED BY USING () / tuple()
# TUPLE IS IMMUTABLE
# TUPLE ALLOWS US TO HAVE DUPLICATE ELEMENTS

# CREATE A TUPLE

t1 = (1,2,3,4,5)
print(type(t1))

t2 = tuple((1,2,3,4,5))
print(type(t2))

t3 = ()
print(type(t3))
```

```python
#SET
#SET NI ADALAH DATA YANG TIDAK TERSUSUN
#SET DONTED BY USING {}
#DOES NOT ALLOW DUPLICATE RECORD
#OUTPUT IS ALL UNIQ IDENTITY NO DUPLICATION
#everobody will get diff result once run

set1={10,20,30,40,50,60,70,80}
print(type(set1))

#set2={(10,20,30,40,50)}
set2=((10,20,30,40,50))  #is set([]) is a function untuk pass nombor kena dua bracket, kalau tak kena guna {}

print(type(set2))

set3=set()
print(type(set3))
```

```python
# no duplicate entry even got duplicate data in set
list1 = [10,20,10,30,40,10,50,60,20,30,50,70]
# WAP to print the no. of unique elements in a List
print('no. of unique elements :',set(list1))
print('no. of unique elements :',len(set(list1)))
```

```python
list1 = [10,20,10,30,40,10,50,60,20,30,50,70]
empty_list = []
for i in list1:                                          #  i --> 30
  if i not in empty_list:                                 # 30 does not exist in []
    empty_list.append(i)                                   # [10,20,30]
print('no. of unique elements :',len(empty_list))
```

```python
# DICTIONARIES
# DICTIONARY USED TO HOLD INFORMATION IN A KEY:VALUE PAIR
# DICTIONARY IS DENOTED BY USING {}
# DICTIONARY IS MUTABLE IN NATURE

emp_info = {'id':101 , 'name':'John' , 'age':32 , 'sal':6700 ,
            'email':'j@gmail.com' , 'city':'klang' , 'dept':'IT'}

print(type(emp_info))
print(emp_info)
print(emp_info["sal"])
print(emp_info["dept"])
print(emp_info.get("name"))
```
