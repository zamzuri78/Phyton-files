```python
# how much on an average churn users spend on phone during day time
avg_min=data[data['churn']==1]['total day minutes'].mean()
print('Churn User Spend During Day tinme:',avg_min)

# how much on an average churn users spend on phone during night time
avg_min=data[data['churn']==1]['total night minutes'].mean()
print('Churn User Spend During Night Time:',avg_min)

#data[data['churn']==1]['total night minutes'].mean())

# how much on an average active users spend on phone during night time
avg_min=data[data['churn']==0]['total night minutes'].mean()
print('Active User Spend During Night Time:',avg_min)
#data[data['churn']==0]['total night minutes'].mean())

# how much on an average churn users spend on phone during evening time
avg_min=data[data['churn']==1]['total eve minutes'].mean()

print('Churn User Spend During Night Time:',avg_min)
#data[data['churn']==1]['total eve minutes'].mean())

# how much on an average active users spend on phone during evening time
avg_min=data[data['churn']==0]['total eve minutes'].mean()
print('Active User Spend Evening Time:',avg_min)
#data[data['churn']==0]['total eve minutes'].mean())
```

```python
data[data['churn']==1].mean(numeric_only=True)

```

```python
data[data['churn']==0].mean(numeric_only=True)
```

```python
# Mean for churned customers
churn_yes = data[data['churn'] == 1].mean(numeric_only=True)

# Mean for non-churned customers
churn_no = data[data['churn'] == 0].mean(numeric_only=True)

# Combine both results into a single comparison DataFrame
compare = pd.DataFrame({
    'Churn = 1 (Yes)': churn_yes,
    'Churn = 0 (No)': churn_no,
    'Difference (Yes - No)': churn_yes - churn_no
})

# Display neatly sorted by difference
compare.sort_values(by='Difference (Yes - No)', ascending=False)
```

```python
#

# Grouping
columns_to_show = ['total day minutes', 'total eve minutes', 'total day minutes']

data.groupby(['churn'])[columns_to_show].describe(percentiles=[])
```
