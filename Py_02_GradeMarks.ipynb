```python
# Loops In Python
# Loops are used to perform a repititive task or used to repeat a block of code

# for loop
# it is used to iterate over a sequence

# AFTER EVERY ITERATION BODY OF FOR LOOP WILL BE EXECUTED

# for iterable_variable in Sequence:
#     code to be executed

product_prices = 10 , 15 , 17 , 19 , 20 , 22 , 26 , 29 , 30 , 32 , 35 , 37 , 40 , 42 , 44 , 45 , 47 , 50 , 52
tax = 1.25
for i in product_prices:
   print(i+tax)
```
