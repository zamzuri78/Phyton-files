```python
#Seaborn - Data visualization library
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import seaborn as sns
import matplotlib.pyplot as plt
#import warnings
#warnings.filterwarnings('ignore')



dataset = sns.load_dataset('iris')
dataset
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
sns.set_style('darkgrid',{'grid.color':'k','grid.linestyle':':'})
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('Line Graph')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
sns.set_style('darkgrid',{'grid.color':'k','grid.linestyle':':'})
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('Line Graph')
plt.show()
```

```python

pokemon = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv')
pokemon
```

```python
#scatterplot()
sns.scatterplot(x='Attack',y='Defense',data=pokemon)
plt.title('Scatter Plot using seaborn')
plt.show()

```

```python
pokemon.Generation.value_counts().sort_index()

sns.scatterplot(x='Attack',y='Defense',hue='Generation',data=pokemon)

plt.title('Count Plot')
plt.show()
```

```python

pkmn_type_colors = ['#3ae40c',  # Grass
                    '#e52106',  # Fire
                    '#0AE3D6',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison

                   ]
sns.scatterplot(x='Attack',y='Defense',hue='Generation',palette=pkmn_type_colors,data=pokemon)
plt.title('Count Plot')
plt.show()
```

```python
pokemon.Generation.value_counts().sort_index()
sns.countplot(x='Generation',data=pokemon)
pokemon.Generation.value_counts().sort_index()
plt.title('Count Plot')
plt.show()
```

```python


# lmplot()
# used to generate a scatter plot using a regression line
colors = ['r', 'y', 'b', 'g', 'c', 'm']

sns.lmplot(x='Attack', y='Defense', hue='Generation', palette=colors, data=pokemon)
plt.title('Scatter Plot Using Seaborn')
plt.show()
```

```python
#countplot - use to generate bar graph

pokemon.Generation.value_counts().sort_index()
plt.figure(figsize=(15,5))
sns.countplot(x='Type 1',hue='Type 1',palette=pkmn_type_colors, data=pokemon)
#sns.countplot(x='Type 1',palette=pkmn_type_colors, data=pokemon)
plt.title('Seaborn Count Plot')
plt.show()
```
