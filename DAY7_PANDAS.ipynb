```python
#in dictionaries  tak boleh panggil array guna indeks number, dalam series boleh
#dictionaries kena assign variable untuk panggil data

#dictionaries ( slicing not allowed) only indexing

import pandas as pd

info = pd.Series(['John', 28, 6700, 'KL'], index=['name', 'age', 'sal', 'city'])
print(info)
print(info['sal'])
print(info.iloc[2])
```

```python
#indexcing slicing and using loops
# Indexing Slicing and Using Loops

s = pd.Series([10,20,30,40,50,60,70] , index=['val1','val2','val3','val4','val5','val6','val7'])

print(s)
 
```

```python
#indexcing in series can be done by using [] ior iloc []
#[] labels
#.loc for both [index Number + labels]
#iloc ---> i stand for index loc----> location
print(s.loc['val1']) #use fpr labels
print(s.iloc[2])  #use for numbers
print(s['val3'])   #use for labels // not advise use this to general easy to trigger error

```

```python
import pandas as pd

# Slicing
#   values : 10  20  30  40  50  60  70
#   index  :  0   1   2   3   4   5   6
#   labels : v1  v2  v3  v4  v5  v6  v7

#s = pd.Series([10, 20, 30, 40, 50, 60, 70],
#              index=['v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'v7'])

# slicing by position
print(s[1:6])
print()

# when slicing by labels, the end label is included
print(s['val2':'val6'])
```

```python
# val1 val4 val7
print(s[['val1','val4','val7']]) #two or double square bracket untuk more than 1 labels
```

```python
# conditional selection
print(s[s % 4 == 0])
```

```python
#using loops
for i in s:
    print(i)
    
```

```python
#guna items
for k,v in s.items():
    print(k,v)
    
```

```python
for i in enumerate(s):  #ini hanya kan beri indeks nombor value untuk key tak akan keluar
    print(i)
    
```
