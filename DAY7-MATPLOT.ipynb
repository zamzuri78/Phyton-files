```python
import matplotlib.pyplot as plt
from matplotlib import style

# style.available  # untuk semak senarai style yang ada
style.use('ggplot')  # guna style ggplot

x1 = [5, 8, 10]
y1 = [12, 16, 6]
x2 = [6, 9, 11]
y2 = [7, 15, 7]
x3 = [11, 13, 12]
y3 = [14, 15, 13]

plt.plot(x1, y1, color='green', linewidth=3, linestyle='dashed', label='Product 1')  # solid / dashed / dotted
plt.plot(x2, y2, color='blue', linewidth=3, linestyle='dashed', label='Product 2')
plt.plot(x3, y3, color='red', linewidth=3, linestyle='dashed', label='Product 2')

plt.title('Graph Using Style')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
plt.legend(loc=2)  # display the labels
plt.grid(True, color='lightblue')

plt.show()
```

```python
import matplotlib.pyplot as plt

# horizontal bar graph
# barh()

products = ['computer', 'laptops', 'mouse', 'keyboards']
quantities = [320, 670, 480, 200]

plt.barh(products, quantities)
plt.xlabel('Quantity')
plt.ylabel('Products')
plt.title('Stock Inventory')
plt.show()
```

```python
import matplotlib.pyplot as plt

# pie plot
activities = ['playing', 'eating', 'working', 'sleeping', 'other']
hrs = [2, 1, 13, 7, 1]
col = ['green', 'cyan', 'orange', 'yellow', 'magenta']

plt.pie(hrs, labels=activities, colors=col)
plt.title('Pie Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# pie plot
activities = ['playing', 'eating', 'working', 'sleeping', 'other']
hrs = [2, 1, 13, 7, 1]
col = ['green', 'cyan', 'orange', 'lightblue', 'magenta']

plt.pie(hrs, labels=activities, colors=col,
        startangle=90,
        explode=(0, 0, 0, 0, 0.1))
plt.pie(hrs, labels=activities, colors=col,
        startangle=90, explode=(0,0,0,0,0.1),
        autopct='%1.1f%%')

plt.title('Pie Plot')
plt.show()
```
